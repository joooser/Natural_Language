```python
import nltk
nltk.download("cess_esp")
```

```python
import re
corpus = nltk.corpus.cess_esp.sents()
print(corpus)
```

```python
flatten = [w for l in corpus for w in l]
print(flatten[:20])
```

```python
arr = {w for w in flatten if re.search("es", w)}
print(arr)
```

```python
arr = [w for w in flatten if re.search("es$", w)]
print(arr)
```

```python
arr = [w for w in flatten if re.search("^es", w)]
print(arr)
```

```python
#rango [a-z] [afr]
arr = [w for w in flatten if re.search("^[afr]", w)]
print(arr[:10])
```
